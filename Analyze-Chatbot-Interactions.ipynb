```python
import pandas as pd
import unidecode
import re
```

```python
df = pd.read_csv("Dados_FAQ.csv")
```

```python
df
```

```python
df[['origem','pergunta','resposta']] = df['Dados'].str.split('|',expand=True)
df.drop(["Dados"],axis=1,inplace=True)
```

```python
df
```

```python
pattern = re.compile("^(##[A-Z]\.[0-9]\.[0-9])")
def limpa(x):
    for i in range(len(x)):
        x[i] = unidecode.unidecode(x[i]).strip()
        if i==2:
            if pattern.match(x[i]):
                return x[i][:7]
            else:
                return None
#     return x[0],x[1],x[2]
df["redirecionamento"] = df.apply(lambda x: limpa(x), axis=1)
df
```

```python
origem = df.origem.value_counts().to_frame().reset_index(level=0)
origem
```

```python
pergunta = df.pergunta.value_counts().to_frame().reset_index(level=0)
pergunta
```

```python
resposta = df[df.redirecionamento.isnull()].resposta.value_counts().to_frame().reset_index(level=0)
resposta
```

```python
redirecionamento = df.redirecionamento.value_counts().to_frame().reset_index(level=0)
redirecionamento
```
